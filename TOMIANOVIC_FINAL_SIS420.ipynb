```python
# ALGORITMO GENETICO BUSCA EL MAYOR VALOR POSIBLE SIN EXCEDER EL PESO
import random

#OPERACIONES DEL ALGORITMO GENETEICO

#un individuo representa una solucion posible
#esta parte del codigo define que tan bueno es el individuo, si es uno lo metera a la mochila, y despues analiza si tiene valor alto
# y su peso no exceda el peso maximo, si el individuo excede el peso maximo este codigo lo sacara de la poblacion o de la mochila

def fitness(individual, items, max_weight): #funcion de aptitud, EVALUA QUE TAN BUENA ES LA SOLUCION(UN INDIVIDUO QUE TAN BUENO ES), de acuerdo si tienen un valor alto sin exceder el peso
    weight, value = 0, 0
    for i in range(len(individual)):
        if individual[i] == 1: # lo metera a la mochila
            weight += items[i][0]
            value += items[i][1]
    if weight > max_weight: #si peso es mayor al permitido la aptitud o solucion es 0, esa solucion no es factible
        return 0
    return value

def mutate(individual): #REALIZA CAMBIOS ALEATORIOS EN UN INDIVIDUO PARA MANTENER DIVERSIDAD , PUEDE VOLVER 0 A 1 ---O--- 1 A 0 SELECCIONANDO UNA PARTE AL AZAR
    index = random.randint(0, len(individual) - 1) #QUE SIGNIFICA LEN
    individual[index] = 1 if individual[index] == 0 else 0

def crossover(parent1, parent2): #CRUCE DE PADRE PARA GENERAR HIJOS
    index = random.randint(1, len(parent1) - 2) # divide a los padres de acuerdo a su tamaño le resta 2 y genera un aleatorio de esa resta, ese sera el tamaño que define como se dividiran y despues uniran los padres
    child1 = parent1[:index] + parent2[index:]
    child2 = parent2[:index] + parent1[index:]
    return child1, child2



# Ejemplo de uUSO
items = [(10, 60), (20, 100),(15, 160), (30, 120),(19, 210),(20, 152)]  # (peso, valor)
max_weight = 60
population_size = 100 #numero de individuos
generations = 100 #numero de veces que repetira el algoritmo

# Inicializar población
population = [[random.randint(0, 1) for _ in range(len(items))] for _ in range(population_size)]# generan 0s y 1s aleatorios, 1 dentro de la mochila 0 fuera de la mochila
#principal
for _ in range(generations):
    population.sort(key=lambda individual: fitness(individual, items, max_weight), reverse=True) #aca se ordenan a los individuos de mayor a menor de acuerdo a su aptitud
    # Seleccionar los mejores individuos para la reproducción
    population = population[:population_size//2] #reduce la poblacion a la mitad manteniendo los mejore individuos
    # Reproducir y mutar
    for i in range(population_size//2, population_size): #recorre la mitad de la poblacion hasta su tamaño completo rellenando la otra mitad
        parent1 = random.choice(population[:population_size//2])
        parent2 = random.choice(population[:population_size//2])
        child1, child2 = crossover(parent1, parent2)
        mutate(child1)
        mutate(child2)
        population.append(child1) #añade a los nuevos hijos a la ootra parte de la poblacion
        population.append(child2)

# Imprimir el mejor individuo
best_individual = max(population, key=lambda individual: fitness(individual, items, max_weight))
print('Mejor individuo:', best_individual, 'Valor:', fitness(best_individual, items, max_weight))
```

```python
import random

def fitness(individual, items, max_weight):
    weight, value = 0, 0
    for i in range(len(individual)):
        if individual[i] == 1: # verifica si la posicion que tomamos es 1, entonces es algo que podemos meter a la mochila
            weight += items[i][0]
            value += items[i][1]
    if weight > max_weight:
        return 0
    return value

def mutate(individual):
    index = random.randint(0, len(individual) - 1)
    old_value = individual[index]
    individual[index] = 1 if individual[index] == 0 else 0
    return old_value, individual[index]

def crossover(parent1, parent2):
    index = random.randint(1, len(parent1) - 2)
    child1 = parent1[:index] + parent2[index:]
    child2 = parent2[:index] + parent1[index:]
    return index, child1, child2

# Ejemplo de uso
items = [(10, 60), (20, 100),(15, 160), (30, 120),(19, 210),(20, 152)]  # (peso, valor)
max_weight = 60
population_size = 100
generations = 100

# Inicializar población
population = [[random.randint(0, 1) for _ in range(len(items))] for _ in range(population_size)]

for g in range(generations):
    population.sort(key=lambda individual: fitness(individual, items, max_weight), reverse=True)
    # Seleccionar los mejores individuos para la reproducción
    population = population[:population_size//2]
    # Reproducir y mutar
    for i in [0, population_size//4, population_size//2 - 1]:  # principio, mitad y final
        parent1 = population[i]
        parent2 = population[(i+1) % (population_size//2)]
        index, child1, child2 = crossover(parent1, parent2)
        old_value1, new_value1 = mutate(child1)
        old_value2, new_value2 = mutate(child2)
        population.append(child1)
        population.append(child2)

# Imprimir el mejor individuo
best_individual = max(population, key=lambda individual: fitness(individual, items, max_weight))

print(f"Generación {g+1}:")
print(f"  Mejor individuo: {population[0]}, Valor: {fitness(population[0], items, max_weight)}")
print(f"  Cruce: {parent1} y {parent2} en el índice {index} producen {child1} y {child2}")
print(f"  Mutación: {child1} cambia el bit {old_value1} a {new_value1}")
print(f"  Mutación: {child2} cambia el bit {old_value2} a {new_value2}")
print('Mejor individuo:', best_individual, 'Valor:', fitness(best_individual, items, max_weight))


```

```python
import numpy as np
import gym #libreria que ayuda a la creacion de de algoritmos de q learning

# Usaremos el entorno Taxi-v3
FoodDelivery = gym.make("Taxi-v3") # el objetivo es recoger pedidos y llevarlos lo antes posible a su destino

# Inicializa la tabla Q con ceros donde el agente aprendera
q_table = np.zeros([FoodDelivery.observation_space.n, FoodDelivery.action_space.n])

# Parámetros de aprendizaje
alpha = 0.9 # learning_rate es cuánto aprende de cada nueva experiencia.
gamma = 0.1 # discount_factore cuánto valora las recompensas si soolo quiere dejar la comida quiere encontrar un mejor camino para dejarla
epsilon = 0.5 # marca el tope de si explora o explota
              #,es cuánto explora o explota de acuerdo a las siguientes partes del codigo acciones aleatorias en lugar de seguir lo que ya sabe.
episodes = 5000 # num_episodes

# Puntos de impresión
print_points = [0, episodes//2, episodes-1] # imprime los episodios principio, mitad y final

for i in range(episodes): # las veces que recorre el escenario # 1 episodio completo es desde que se recoge y se deja la comida en el destino correcto
    state = FoodDelivery.reset() #resetea todo el entorno al estado inicial coloacando  la comida y al delivery en diferente lugar

    done = False #se utiliza para rastrear si el episodio del juego ha terminado o no
    steps = 0
    if i in print_points:
        print(f"\nEpisodio: {i+1}")
    while not done:
        # Elige una acción. Selecciona una acción aleatoria con probabilidad epsilon, de lo contrario, selecciona la mejor acción en la tabla Q
        if np.random.uniform(0, 1) < epsilon: #generamos un numero aleatorio, si es menor a epsilon explorara y aprendera mayores acciones y estrategias
            action = FoodDelivery.action_space.sample() # Exploración: elige una acción aleatoria Esta línea de código se ejecuta si el número aleatorio es menor que epsilon.
                                                            #AQUI ESTAN LAS ACCIONES DEL 0 A 5
        else:
            action = np.argmax(q_table[state]) # Explotación: elige la mejor acción en la tabla Q Esta línea de código se ejecuta si el número aleatorio es mayor o igual que epsilon
                                                #utiliza lo que aprendio para tomar las mejores acciones

#actualización de la tabla Q
        next_state, reward, done, info = FoodDelivery.step(action)  ##Aquí, el agente realiza la acción que eligió en el paso anterior.
          #next_state = es el nuevo estado del entorno después de realizar la acción.
          #reward = es la recompensa que el agente recibe por realizar la acción.
          #done = es un valor booleano que indica si el episodio ha terminado (por ejemplo, si el delivery ha dejado la comida en el destino correcto).

        # Actualiza la tabla Q usando la ecuación de actualización de Q-Learning
        old_value = q_table[state, action] #Aquí, estamos obteniendo el valor antiguo de la tabla Q para la acción que acabamos de realizar en el estado actual.
        next_max = np.max(q_table[next_state]) #representa la mejor acción que podemos tomar en el próximo estado,

        new_value = (1 - alpha) * old_value + alpha * (reward + gamma * next_max) #
        q_table[state, action] = new_value # se actualiza la tabla q con el nuevo valor

        state = next_state
        steps += 1

        # Visualiza el entorno en los puntos de impresión
        if i in print_points:
            print(FoodDelivery.render(mode='ansi')) #ansi grafica el entorno y agente en cadena de texto

    if i in print_points:
        print(f"Número de pasos: {steps}")



    np.save("q_table.npy", q_table)


# Imprime la tabla Q final
print("\nTabla Q final:")
print(q_table)
```
